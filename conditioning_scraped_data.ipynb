```python
import pandas as pd
```

First combine all the stats together

```python
offense = pd.read_csv('offense_results.csv')
defense = pd.read_csv('defense_results.csv')
special = pd.read_csv('special_results.csv')
```

```python
def clean_stats_df(df):
    print(len(df))
    df_new = df.drop(['Unnamed: 0', 'ranker','g'], axis=1).set_index(['school_name','year'])
    return df_new
```

```python
clean_off = clean_stats_df(offense)
clean_def = clean_stats_df(defense)
clean_spec = clean_stats_df(special)
```

```python
all_stats = pd.merge(clean_off, clean_def, how='left', left_index=True, right_index=True)
all_stats = pd.merge(all_stats, clean_spec, how='left', left_index=True, right_index=True)
```

```python
all_stats.info()
```

Now bring in the conference standings and final ranks.

```python
standings = pd.read_csv('standings_results.csv')
```

```python
def clean_standings_df(df):
    print(len(df))
    df_new = df.drop(['ranker'], axis=1).set_index(['school_name','year'])
    return df_new
```

```python
clean_stand = clean_standings_df(standings)
```

```python
all_stats = pd.merge(all_stats, clean_stand, how='left', left_index=True, right_index=True)
```
